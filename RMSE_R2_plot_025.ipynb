```python
import matplotlib as mpl 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
```

```python
x = ['All Features ' , "Without Latitude" , "Without Latitude and Commuted Density"]
yTraining_RMSE = [0.32615,0.34804,0.36894] 
err_yTraining_RMSE = [0.01566 ,0.01628 ,0.01789]
yTest_RMSE = [0.90023,0.96245,1.01980 ]
err_yTest_RMSE = [0.11901 ,0.12078 , 0.13442 ]
yTraining_R2 = [0.94754 ,0.94015 ,0.93286 ]
err_yTraining_R2 = [0.00447 ,0.00511,0.00533 ]
yTest_R2 = [0.58379,0.52817,0.46673]
err_yTesr_R2 = [0.09836 , 0.10710 ,0.11597]
fig , ax = plt.subplots( figsize = (15,10))
ax.set_title('Andamento RMSE Test Split 0.30')
ax.set_xlabel('Features')
ax.set_ylabel('RMSE')
plt.plot(x , yTraining_RMSE ,'ro' , label = 'Training RMSE')
ax.errorbar(x , yTraining_RMSE , err_yTraining_RMSE , fmt = 'none' , ecolor= 'r' , linewidth = 1 , capsize = 6)
plt.plot(x , yTest_RMSE , 'bo' , label = 'Test RMSE')
ax.errorbar(x , yTest_RMSE , err_yTest_RMSE , fmt = 'none' , color = 'b' , linewidth = 1 , capsize=6)
plt.legend()
```
