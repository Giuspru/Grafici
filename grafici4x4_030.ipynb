```python
plt.subplots(figsize = (23 ,15))

plt.subplot(2,2,1)
y = df1Training[' RMSE Seconda Posizione ']
z = df1Training[' RMSE prima Posizione ']
x = df1Training["         Features "]
X_axis = np.arange(len(x))
plt.xticks(X_axis , x )
plt.yticks(np.arange(0 , 150 , 5))
plt.bar(X_axis  , y , 0.4 , label = 'Seconda Posizione' , color = 'orange'  )
plt.bar(X_axis , z , 0.4 , label = 'Prima Posizione' , color = 'blue')
plt.xticks(rotation = 45)
plt.xticks(ha = 'center')
plt.xlabel("Features")
plt.ylabel("% Importance of feature")
plt.title('Training RMSE 0.30 all features')
plt.legend()

plt.subplot(2,2,2) #Questo funziona 
y = df1Test[' RMSE Seconda Posizione ']
z = df1Test[' RMSE prima Posizione ']
x = df1Test["            Features "]
X_axis = np.arange(len(x))
plt.xticks(X_axis , x )
plt.yticks(np.arange(0 , 150 , 5))
plt.bar(X_axis + 0.2 , y , 0.4 ,   label = 'Seconda Posizione' , color = 'orange')
plt.bar(X_axis - 0.2 , z  , 0.4 ,  label = 'Prima Posizione' , color = 'blue')
plt.xticks(rotation = 45)
plt.xticks(ha = 'center')
plt.xlabel("Features")
plt.ylabel("% Importance of feature")
plt.title('Test RMSE 0.30 all features')
plt.legend()

plt.subplot(2,2,3)
y = df1Training[' R2 Prima Posizione ']
z = df1Training[' R2 Seconda Posizione ']
x = df1Training["         Features "]
X_axis = np.arange(len(x))
plt.xticks(X_axis , x )
plt.xticks(X_axis , x )
plt.yticks(np.arange(0 , 150 , 5))
plt.bar(X_axis  , y , 0.4 ,label = 'Seconda Posizione' , color = 'orange'  )
plt.bar(X_axis , z , 0.4 , label = 'Prima Posizione' , color = 'blue')
plt.xticks(rotation = 45)
plt.xticks(ha = 'center')
plt.xlabel("Features")
plt.ylabel("% Importance of feature")
plt.title('Training R2 0.30 all features')
plt.legend()

plt.subplot(2,2,4)
y = df1Test[' R2 Prima Posizione ']
z = df1Test[' R2 Seconda Posizione ']
x = df1Test["            Features "]
X_axis = np.arange(len(x))
plt.xticks(X_axis , x )
plt.yticks(np.arange(0 , 150 , 5))
plt.bar(X_axis  , y , 0.4, label = 'Seconda Posizione' , color = 'orange'  )
plt.bar(X_axis , z , 0.4 ,  label = 'Prima Posizione' , color = 'blue')
plt.xticks(rotation = 45)
plt.xticks(ha = 'center')
plt.xlabel("Features")
plt.ylabel("% Importance of feature")
plt.title('Test R2 0.30 all features')
plt.legend()

plt.tight_layout()


```
