```python
y = df1Training[' RMSE Seconda Posizione ']
z = df1Training[' RMSE prima Posizione ']
x = df1Training['Features ']
fig , ax  = plt.subplots(figsize=(16 , 5) )
X_axis = np.arange(len(x))
plt.xticks(X_axis , x )
plt.yticks(np.arange(0 , 150 , 5))
plt.bar(X_axis + 0.2 , y , 0.4 , label = 'Seconda Posizione' , color = 'orange')
plt.bar(X_axis - 0.2 , z , 0.4 , label = 'Prima Posizione' , color = 'blue')
plt.xlabel("Features")
plt.ylabel("% Importance of feature")
plt.title('Training RMSE 0.25 all features')
plt.legend()
```
