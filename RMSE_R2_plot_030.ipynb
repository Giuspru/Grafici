```python
import matplotlib as mpl 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
```

```python
#just let's retrive some data:
x = ['All Features ' , "Without Latitude" , "Without Latitude and Commuted Density"]
yTraining_RMSE = [0.33031 , 0.35147 , 0.37243] 
err_yTraining_RMSE = [0.01850, 0.02217 , 0.01989 ]
yTest_RMSE = [0.88883,0.94614, 1.00648]
err_yTest_RMSE = [0.10519 ,0.12216, 0.11670  ]
yTraining_R2 = [0.94637,0.93928 ,0.93190]
err_yTraining_R2 = [0.00501 ,0.00586,0.00667]
yTest_R2 = [0.59501,0.54510,0.48029]
err_yTesr_R2 = [0.08034 ,0.08892 ,0.10325 ]
fig , ax = plt.subplots( figsize = (15,10))
ax.set_title('Andamento RMSE Test Split 0.30')
ax.set_xlabel('Features')
ax.set_ylabel('RMSE')
plt.plot(x , yTraining_RMSE ,'ro' , label = 'Training RMSE')
ax.errorbar(x , yTraining_RMSE , err_yTraining_RMSE , fmt = 'none' , ecolor= 'r' , linewidth = 1 , capsize = 6)
plt.plot(x , yTest_RMSE , 'bo' , label = 'Test RMSE')
ax.errorbar(x , yTest_RMSE , err_yTest_RMSE , fmt = 'none' , color = 'b' , linewidth = 1 , capsize=6)
plt.legend()


```
